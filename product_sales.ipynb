```python
import pandas as pd
import numpy as np
import matplotlib.style as style
import matplotlib.pyplot as plt
```

```python
# reading in the files 
sales_df = pd.read_csv("./product_sales.csv", skipinitialspace= True)
sales_df.head() 
```

```python
# checking the datatype of each column
sales_df.dtypes 
```

```python
# changing the datatype 
sales_df['Date'] = sales_df['Date'].astype('datetime64[ns]')
sales_df.dtypes 
```

```python
sales_df.head()
```

```python
# checking for null values
sales_df.isna().sum()
```

```python
sales_df.columns
```

```python
# changing a column name
sales_df.rename(columns={"Age_Group_NUMBER":"Age_Group_Number"}, inplace= True)
sales_df.head()
```

```python
# checking for wrongly spelt values
sales_df["Country"].unique() 
```

```python
sales_df["Country"].replace({"canada":"Canada"}, inplace= True)
sales_df.head()
```

```python
# checking for wrongly spelt values
sales_df["Customer_Gender"].unique()
```

```python
# checking for wrongly spelt values
sales_df["Age_Group"].unique()
```

```python
# checking for wrongly spelt values
sales_df["Product_Category"].unique() 
```

```python
# Analysis
grp1 = sales_df.groupby(["Age_Group"])
df2 = grp1["Profit"].sum().reset_index().rename(columns={'Profit': 'Profit'})
df2 = df2.sort_values(by= "Profit", ascending= True)
df2.head()
```

```python
# customer segmentation by profit
style.use("fivethirtyeight")
a = df2["Age_Group"].tolist()
b = df2["Profit"].tolist()
plt.barh(a, b, color = "g")
plt.ylabel("Age Group")
plt.xlabel("Profit")
plt.title("Customer Segmentation by Profit")
plt.show() 
```

```python
# Analysis
grp2 = sales_df.groupby(["Product_Category"])
df3 = grp2["Order_Quantity"].sum().reset_index().rename(columns={'Order_Quantity': 'Order_Quantity'})
df3 = df3.sort_values(by= "Order_Quantity", ascending= True)
df3.head() 
```

```python
# Order Quantity Count Analysis
style.use("fivethirtyeight")
c = df3["Product_Category"].tolist()
d = df3["Order_Quantity"].tolist()
plt.bar(c, d, color = "g")
plt.ylabel("Order Quantity")
plt.xlabel("Product Category")
plt.title("Order Quantity Count")
plt.show() 
```
